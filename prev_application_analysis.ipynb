### Data Cleaning

```python
def missingdata_percentage(df):
    missing_df = pd.DataFrame(columns=['category','percentage'])
    for col in df.columns:
        if df[col].isnull().values.any():
            percentage = df[col].isnull().sum()/len(df[col])*100
            missing_df = missing_df.append({'category':col,'percentage':percentage},ignore_index=True)
    return missing_df
```

```python
import warnings
warnings.filterwarnings('ignore')

prev_df_missingData = missingdata_percentage(prev_df)
prev_df_missingData.sort_values(by='percentage',ascending=False)
```

- Observation: There are 16 features in the Prev_df which are having the missing values.

```python
def value_wise_default_percentage(df,col):
    new_df = pd.DataFrame(columns=['Value','Percentage of Defaulter'])
    
    for value in df[col].unique():
        default_count = df[(df[col]==value)&(df['TARGET']==1)].shape[0]
        total_count = df[df[col]==value].shape[0]
        new_df = new_df.append({'Value':value,'Percentage of Defaulter':(default_count*100/total_count)},ignore_index=True)
    
    return new_df.sort_values(by='Percentage of Defaulter',ascending=False)
```

```python
value_wise_default_percentage(prev_df_merged,'NAME_GOODS_CATEGORY').unstack().fillna(0)*100
```
